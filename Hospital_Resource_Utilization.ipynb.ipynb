```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
patients = pd.DataFrame({
    "patient_id": range(1, 101),
    "age": np.random.randint(18, 80, 100),
    "gender": np.random.choice(["Male", "Female"], 100),
    "insurance": np.random.choice(["Private", "Government", "Self-Pay"], 100)
})

admissions = pd.DataFrame({
    "patient_id": range(1, 101),
    "department": np.random.choice(
        ["Cardiology", "Orthopedics", "General Medicine", "Emergency", "Pediatrics"], 100),
    "length_of_stay": np.random.randint(1, 15, 100),
    "emergency_case": np.random.choice(["Yes", "No"], 100)
})

patients.head(), admissions.head()
```

```python
alos = admissions["length_of_stay"].mean()

emergency_split = admissions["emergency_case"].value_counts()

dept_admissions = admissions["department"].value_counts()

alos, emergency_split, dept_admissions
```

```python
plt.figure()
dept_admissions.plot(kind="bar")
plt.title("Admissions by Department")
plt.xlabel("Department")
plt.ylabel("Number of Patients")
plt.tight_layout()
plt.savefig("admissions_by_department.png")
plt.show()
```

```python
plt.figure()
emergency_split.plot(kind="pie", autopct="%1.1f%%")
plt.title("Emergency vs Scheduled Cases")
plt.ylabel("")
plt.tight_layout()
plt.savefig("emergency_vs_scheduled.png")
plt.show()
```

```python
plt.figure()
plt.hist(admissions["length_of_stay"], bins=10)
plt.title("Length of Stay Distribution")
plt.xlabel("Days")
plt.ylabel("Patients")
plt.tight_layout()
plt.savefig("length_of_stay.png")
plt.show()
```

```python
patients.to_csv("patients.csv", index=False)
admissions.to_csv("admissions.csv", index=False)
```

```python
print("Key Insights:")
print(f"- Average Length of Stay: {alos:.2f} days")
print(f"- Department with highest admissions: {dept_admissions.idxmax()}")
print(f"- Emergency cases percentage: {(emergency_split['Yes']/len(admissions))*100:.1f}%")
```
